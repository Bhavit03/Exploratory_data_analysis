```python
df_country=pd.read_excel('Country-Code.xlsx')
```

```python
df_country.head()
```

```python
final_df=pd.merge(df,df_country,how='left',on='Country Code')
final_df.head()
```

```python
final_df.dtypes
```

```python
country_names=final_df.Country.value_counts().index
country_names
```

```python
country_values=final_df.Country.value_counts().values
```

```python
# PIE CHART - TOP 3 COUNTRIES THAT USES ZOMATO BASED ON ABOVE DATA
plt.pie(country_values[:3],labels=country_names[:3],autopct="%1.2f%%")
```

OBSERVATION: MAXIMUM TRANSACTIONS / ORDERS OF ZOMATO ARE FROM INDIA, THEN USA ,THEN UK.

```python
final_df.columns
```

```python
ratings=final_df.groupby([ 'Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

# OBSERVATIONS
1. WHEN RATING IS FROM 1.8 - 2.4------> POOR
2. WHEN RATING IS FROM 2.5 - 3.4------> AVERAGE
3. WHEN RATING IS FROM 3.5 - 3.9------> GOOD
4. WHEN RATING IS FROM 4.0 - 4.4------> VERY GOOD
5. WHEN RATING IS FROM 4.5 - 4.9------> EXCELLENT

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(y='Rating count',x='Aggregate rating',hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

```python
# FIND THE COUNTRY THAT HAS GIVEN 0 RATING
final_df[final_df['Aggregate rating']==0].groupby('Country').size().reset_index()
```

```python
# FIND OUT WHICH CURRENCY IS USED BY WHICH COUNTRY
final_df[['Country','Currency']].groupby(['Country','Currency']).size().reset_index()
```
